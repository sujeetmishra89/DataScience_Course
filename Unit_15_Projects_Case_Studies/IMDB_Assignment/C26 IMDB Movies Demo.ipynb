```python
df1 = ipl[['team1', 'team2', 'win_by_runs', 'win_by_wickets']]
df1.head()
```

```python
# Add cnt column with values as 1

df1['cnt']=1
```

```python
df1.head()
```

```python
# Group by - t1 and t2
t1 = df1.groupby('team1').sum()
t1.head()
```

```python
t2 = df1.groupby('team2').sum()
t2.head()
```

```python
# Add t1 and t2

t3 = t1.add(t2, fill_value=0)
```

```python
t3.head()
```

```python
t3.iloc[:, 0:-1] = t3.iloc[:, 0:-1].divide(t3.cnt, axis = 0)
```

```python
t3.head()
```

```python
t3.iloc[:, 0] = t3.iloc[:, 0].astype(int)
```

```python
t3.head()
```

```python
t3.iloc[:, 1] = round(t3.iloc[:, 1], 2)
```

```python
t3.head()
```

```python
t3.sort_values(by = 'cnt', ascending = False, inplace = True)
```

```python
plt.figure(figsize = (20,5))
sns.barplot(x = t3.index.values, y = t3.cnt, color = 'r')
plt.xticks(rotation = 90)
plt.xlabel('Teams')
```

```python
plt.figure(figsize = (20,10))
plt.subplot(1,2,1)
sns.heatmap(t3[['win_by_runs', 'cnt']], annot = True, fmt = 'd')
plt.xticks(rotation = 90)

plt.subplot(1,2,2)
sns.heatmap(t3[['win_by_wickets', 'cnt']], annot = True, fmt = 'f')
```
