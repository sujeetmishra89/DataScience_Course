```python
# 28

df_by_genre = ipl[ ['team1','team2','win_by_runs','win_by_wickets'] ]
df_by_genre.head()
```

```python
# 29

df_by_genre['cnt'] = 1
```

```python
df_by_genre.head()
```

```python
# 30

g1 = df_by_genre.groupby('team1').sum()
g1
```

```python
g2 = df_by_genre.groupby('team2').sum()
g2
```

```python
# 31

df_add = g1.add(g2, fill_value=0)
df_add
```

```python
# 32

genre_top10 = df_add[ df_add['cnt']>=100 ]
genre_top10
```

```python
# 33

genre_top10.iloc[ :,:-1 ]
```

```python
genre_top10.iloc[ :,:-1 ] = genre_top10.iloc[ :,:-1 ].divide(genre_top10['cnt'], axis=0)
```

```python
genre_top10
```

```python
# 34
# Converting CVotes to int type

genre_top10.iloc[:,0] = genre_top10.iloc[:,0].astype(int)
```

```python
genre_top10
```

```python
# 35
# Rounding off the columns of Votes to two decimals

genre_top10.iloc[:,1] = round(genre_top10.iloc[:,1], 2)
```

```python
genre_top10
```

```python
genre_top10.index
```

```python
# 36

plt.figure(figsize=(20,5))
sns.barplot(x = genre_top10.index, y=genre_top10['cnt'])
# plt.xticks(rotation=45)
plt.show();
```
